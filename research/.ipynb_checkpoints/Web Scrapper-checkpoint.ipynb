# Web Scrapper code


```python
import re
import os
import logging
import time
import datetime
import zipfile
from urllib.request import urlopen as ureq
from bs4 import BeautifulSoup as soup
from lxml import html, etree
from boto.s3.key import Key
import sys
```

```python
base_url = 'https://www.azlyrics.com'
artist_index = "a"
url_rendered = base_url + '/' + artist_index +'.html'
print(url_rendered)
```

```python
uCLient = ureq(url_rendered)
page_html=uCLient.read()
```

```python
page_soup = soup(page_html, 'html.parser')
```

```python
print(page_soup.prettify())
```

```python
divs = page_soup.find_all("div", {"class": "artist-col"})
print(divs)
```

```python
artistarray = []
for div in divs:
    all_href = div.find_all("a")
#     print(all_href)
    for link in all_href:
        a = [link.text,link['href']]
        artistarray.append(a)
print(artistarray[0][1])
```

```python
artist_url = base_url + '/' +artistarray[0][1]
print(artist_url)
uCLient = ureq(artist_url)
page_html=uCLient.read()
```

```python
page_soup = soup(page_html, 'html.parser')
div = page_soup.find("div", {"id": "listAlbum"})
print(div)
```
